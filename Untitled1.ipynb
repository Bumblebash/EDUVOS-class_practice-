```python
def ticket_price():
    user_input1 = 0
    user_input2 = 0
    Balance = 0
    ticket_price = float(45.5)
    while True:
        inp1 = input("Enter Number of Students")
        inp2 = input("Enter Ammount")
        inp3 = input("Is that All?")
        try:
            user_input1 = float(inp1)
        except ValueError:
            print("Enter a Valid Numerical Value")
        try:
            user_input3 = str(inp3)
        except:
            print("Enter a Valid Respose(done)")
        try:
            user_input2 = float(inp2)
        except:
            print("Enter a Valid Numerical Value")
        if inp3.lower() == 'done':
            break
    total_cost = user_input1 * ticket_price
    possible_tickets = user_input2 / ticket_price
    Balance = float(user_input2 - total_cost)
    return total_cost, user_input1, possible_tickets, Balance
total_cost, user_input1, possible_tickets, Balance = ticket_price()
print("Number of Learners / Tickets Sold:", user_input1)
print("Total_cost/Money_spent:", f"R{total_cost}")
print("Possible Number of Tickets", int(possible_tickets))
print("Balance/Change:", f"R{Balance}")
```
