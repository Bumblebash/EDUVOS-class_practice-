```python
user_input = []
count = 0
total = 0
while True:
    inp = input("Enter a number")
    if inp.lower() == "done":
        break
    try:
        number = float(inp)
        user_input.append(number)
    except ValueError:
        print("Enter a Valid Number")
    count +=1 
    total += number
average = total / count
print("user_input", user_input)
print("count", count)
print("sum", total )
print("average", average)
        
        
        
    
        
```
