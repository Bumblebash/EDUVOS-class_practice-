## Exercise 3;
- Write a program(Python Function) that determines wether you can run for president.
- To run for president, the constitution states that:
- No person except a natural born citizen, of the UnitedStates, at the time of adoption of this Constitution shall be eligible to the office of President, neither shall be eligible to that office who shall not have
- shall not have attained to the Age of thirty five years(35 years), and been Fourteen Years a Resident within the United states(14 years).
- Ask three questions of the user and use the guess and Check Pattern to dtermine if they are eligible to run for president.
- Assuming the ammendment was valid in 1788

```python
#Conditions
#Natural Born Citizen of the US at the time of adoptiion of this Constituion
#35 years of Age (Minimum)
# 14 years residency(Minimum)
```

```python
def eligibility():
    #Checking for Applicant Citizenship
    while True:
        citizenship = input("What is your Citizenship?: ").strip()
        if not citizenship:
            print("Input cannot be left Blank. Please enter your citizenship.")
        else:
            break
    #Checking for Applicant Citizenship Type
    while True:
        citizenship_type = input("What is your citizenship Type?(Birth, Derivation, Naturalization, Acquisition: ").strip()
        if not citizenship_type:
            print("Input cannot be left Blank, Please enter  your citizenship type")
        else:
            break
    #Checking for Applicant Age
    while True:
        try:
            age = int(input("What is your Age?: ").strip())
            break
        except ValueError:
            print("Enter a Numerical Value")
    #Checking for Applicant Period of Stay
    while True:
        try:
            period_of_stay = float(input("How long have you been in the US(Years)?: "))
            break
        except ValueError:
            print("Enter a period in approximate years")
    #Relevant Expected User Inputs
    valid_countries = ["USA", "US", "UNITED STATES OF AMERICA", "UNITED STATES"]
    if (citizenship.upper() in valid_countries and citizenship_type.lower() == "birth" and
       age >= 35 and 
       period_of_stay >= 14 ):
        #Positive Result
        print("\n ====== Congratulations!!!======")
        print(" Based on your answers ,You  meet the constitutional requiremnents  to run for presidency!")
    else:
        #Unfortunate Result
        print("Unfortunately , You dont meet the criteria for running for presidency")
eligibility()



            
    
        
        
```

```python
def check_eligibility():
    # Check US Citizenship
    while True:
        is_natural_born = input("Are you a natural-born citizen of the United States? (yes/no): ").strip().lower()
        if not is_natural_born:
            print("Field cannot be left blank, Kindly confirm  your citizenship type")
        else:
            break
    #Check Applicant Age
    while True:
        try:
            applicant_age = int(input("Enter your age?: ").strip())
            break
        except ValueError:
            print("Enter a Valid Numerical Value")
    #Years of Residency in the US
    while True:
        try:
            years_of_residency = float(input("How many years have you been a resident in the United States?: ").strip())
            break
        except ValueError:
            print("Enter a valid Nuerical value(Approximate years in the US)")
    # Checking and  Validating the constitutional requirements 
    if is_natural_born == "yes" and applicant_age >= 35 and years_of_residency >= 14:
        print("\n ===Eligibility Result ====")
        print("Congratulations! Based on your answers, you meet the constitutional requirements to run for US President.")
    else:
        print("\n ===Eligibility Result====")
        print("Sorry, you do not meet the constitutional requirements to run for US President.")

# Run the program
if __name__ == "__main__":
    check_eligibility()
```
