```python
x = float(input("please enter the number"))
y = float(input("2nd number"))

print("Add the two numbers")
print("Subtract the two numbers")
print("Multiply the two numbers")
print("Divide the two numbers")

choice = int(input("Please enter your choice:"))
print("The answer is:", end ="")
if choice == 1:
    print (x + y)
else:
    if choice == 2:
        print(x - y)
    else :
        if choice == 3:
            print(x * y)
        else :
            if choice == 4:
                print(x / y)
            else:
                print("You did not enter a valid choice.")
```

### Exercise 2
- Use a the guess and check pattern to determine if a triangle is  a Perfect triangle.
- A perfect triangle has side lenghts tat are mulltiples of 3 and 4 and 5
- Ask the usre to enter the shortest, muiddle, and longest sides of a triangle
- and then print "It is a perfect triangle", "If its is" and "It is not a perfevt triangle", if itsn't
- You may assume that the side lenghts are integers
- Let your guess be that the message you will print is "It is aperfect triangle"

```python
def triangle():
    while True:
        try:
            a = float(input("Enter a  short side:"))
            break
        except ValueError:
            print("Enter a Valid float value")
    while True:
        try:
            b = float(input("Enter a middle side:"))
            break
        except ValueError:
            print("Enter a valid float value")
    while True:
        try:
            c  = float(input("Enter a long side"))
            break
        except ValueError:
            print("Enter a valid float value")
    if(a**2 + b**2 == c**2):
        print("It is a perfect Triangle")
    else:
        print("Not a perfect Triangle")
triangle()


```

## Exercise 3;
- Write a program(Python Function) that determines wether you can run for president.
- To run for president, the constitution states that:
- No person except a natural born citizen, of the UnitedStates, at the time of adoption of this Constitution shall be eligible to the office of President, neither shall be eligible to that office who shall not have
- shall not have attained to the Age of thirty five years(35 years), and been Fourteen Years a Resident within the United states(14 years).
- Ask three questions of the user and use the guess and Check Pattern to dtermine if they are eligible to run for president.
- Assuming the ammendment was valid in 1788

```python
#Conditions
#Natural Born Citizen of the US at the time of adoptiion of this Constituion
#35 years of Age (Minimum)
# 14 years residency(Minimum)
```

```python
def validity():
    while True:
        try:
            citizenship = str(input("What is your Citizenship?").strip())
        except ValueError:
            print("Enter a valid Citizenship")
        if not citizenship:
            print("Input cannot be left Blank. Please enter your citizenship.")
        else:
            break
    while True:
        try:
            citizenship_type = str(input("What is your citizenship Type?(Birth, Derivation, Naturalization, Acquisition, ")).strip()
        except ValueError:
            print("Enter Valid citizenhip Options")
        if not citizenship_type:
            print("Input cannot be lest Blank, Please enter  your citizenship type")
        else:
            break
    while True:
        try:
            age = int(input("What is your Age?").strip())
            break
        except ValueError:
            print("Enter a Numerical Value")
    while True:
        try:
            period_of_stay = float(input("How long have you been in the US(Years)?"))
            break
        except ValueError:
            print("Enter a period in approximate years")
    valid_countries = ["USA", "US", "UNITED STATES OF AMERICA", "UNITED STATES"]
    if (citizenship.upper() in valid_countries and citizenship_type.lower() == "birth" and
       age >= 35 and 
       period_of_stay >= 14 ):
        print("\n You are eligible to run for presidency!")
    else:
        print("Unfortunately , You dont meet the criteria for running for presidency")
validity()



            
    
        
        
```

```python
def check_eligibility():
    # 1. Ask three questions of the user
    is_natural_born = input("Are you a natural-born citizen of the United States? (yes/no): ").strip().lower()
    age_input = int(input("What is your current age? (e.g., 35): ").strip())
    years_resident = int(input("How many years have you been a resident in the United States? (e.g., 14): ").strip())

    # 2. Guess and Check Pattern: Define criteria and evaluate input
    # Guess: Assume the user is NOT eligible by default
    is_eligible = False

    # Check: Validate the constitutional requirements
    if is_natural_born == "yes" and age_input >= 35 and years_resident >= 14:
        is_eligible = True

    # 3. Output the result
    print("\n--- Eligibility Result ---")
    if is_eligible:
        print("Congratulations! Based on your answers, you meet the constitutional requirements to run for US President.")
    else:
        print("Sorry, you do not meet the constitutional requirements to run for US President.")

# Run the program
if __name__ == "__main__":
    check_eligibility()
```
